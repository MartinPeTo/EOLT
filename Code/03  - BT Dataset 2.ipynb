### BT Dataset management

From the full BT dataset we have to clean the data in order to get the following characteristics:

    -Only one row per SN. (full SN required)
    
    -¿Keep the last "pass" test of each drawer?.
    
    -Not null data.

```python
#Import:
import numpy as np
import pandas as pd
```

```python
bt_df = pd.read_csv('../Data/BT_Report.csv')
```

```python
#deleting the ID column:.

bt_df = bt_df.drop("ID", axis = 1)

#We delete all the RESULTNUM columns:
import re

columns = bt_df.columns
for col in columns:
    if re.match(r'^RESULTN', col):
        bt_df.drop(col, axis=1, inplace=True)
```

```python
#Renaming all the columns:
bt_columns = ['SN','TEST','day','time','TESTVERSION','status','FIRST','LAST','SHIFT',
             'FAMILY','Calibration','@TABLE','RADIUS_FRONT_BIG_ROLL','BTFrontBigError_25N',
              'BTFrontBigError_20N','BTFrontBigError_15N','BTFrontBigError_30N',
             'AccBTFrontBigError_25N','DecBTFrontBigError_30N','RADIUS_FRONT_SMALL_ROLL',
             'BTFrontSmallError_25N','BTFrontSmallError_20N','BTFrontSmallError_15N',
             'BTFrontSmallError_30N','AccBTFrontSmallError_25N','DecBTFrontSmallError_30N',
             'RADIUS_REAR_BIG_ROLL','BTRearBigError_25N','BTBTRearBigError_20N',
             'BTBTRearBigError_15N','BTBTRearBigError_30N','AccBTRearBigError_25N',
             'DecBTRearBigError_30N','RADIUS_REAR_SMALL_ROLL','BTRearSmallError_25N',
             'BTRearSmallError_20N','BTRearSmallError_15N','BTRearSmallError_30N',
             'AccBTRearSmallError_25N','DecBTRearSmallError_30N']
bt_df.columns = bt_columns
bt_df.head()
```

### Cleaning dataset

Columns TESTVERSION, FIRST, LAST, SHIFT, FAMILY, FAMILY, Calibreation, @TABLE are not relevant for the analysis.

```python
#Deleting the not useful columns:

drop_columns2 = ['TEST','TESTVERSION','FIRST','LAST','SHIFT','FAMILY','Calibration','@TABLE']
bt_df.drop(drop_columns2, axis=1, inplace=True)
```

```python
#SN has a format that is DRWXXXXXXXXX. We remove rows with other SN:

bt_df = bt_df[bt_df.SN != 'DRW']
```

We must unify the DAY and TIME column into a datetime column to filter each SN by the last test:

```python
#We create a datetime column that concatenate both day and time columns into only one:
#The column time we ensure to have 6 elements adding a 0 in case it didn't.
# convert_time = lambda x: str(x) if len(str(x))==6 else ('0' + str(x))

def convert_time(x):
    while len(str(x)) < 6:
        x = '0' + str(x)
    return str(x)

bt_df['datetime'] = bt_df['day'].map(str) + bt_df['time'].apply(convert_time)
bt_df['datetime'] = pd.to_numeric(bt_df['datetime'])
```

```python
#We reorder the columns:

bt_df = bt_df[['SN','day','time','datetime','status','RADIUS_FRONT_BIG_ROLL','BTFrontBigError_25N',
              'BTFrontBigError_20N','BTFrontBigError_15N','BTFrontBigError_30N',
             'AccBTFrontBigError_25N','DecBTFrontBigError_30N','RADIUS_FRONT_SMALL_ROLL',
             'BTFrontSmallError_25N','BTFrontSmallError_20N','BTFrontSmallError_15N',
             'BTFrontSmallError_30N','AccBTFrontSmallError_25N','DecBTFrontSmallError_30N',
             'RADIUS_REAR_BIG_ROLL','BTRearBigError_25N','BTBTRearBigError_20N',
             'BTBTRearBigError_15N','BTBTRearBigError_30N','AccBTRearBigError_25N',
             'DecBTRearBigError_30N','RADIUS_REAR_SMALL_ROLL','BTRearSmallError_25N',
             'BTRearSmallError_20N','BTRearSmallError_15N','BTRearSmallError_30N',
             'AccBTRearSmallError_25N','DecBTRearSmallError_30N']]
```

The test columns are divided in two: test name and test result. We delete the first ones and rename the second ones.

```python
#Groupby SN and select last test
#https://stackoverflow.com/questions/41525911/group-by-pandas-dataframe-and-select-latest-in-each-group

bt_df = bt_df.sort_values('datetime').groupby('SN').tail(1)

bt_df.head()
```
