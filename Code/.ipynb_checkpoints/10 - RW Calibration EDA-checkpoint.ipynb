```python
#Import

import pandas as pd
import numpy as np

###Â VISUALIZATION ###
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
rw = pd.read_csv('../Data/RW.csv')
rw.head()
```

### EDA Analysis

```python
#We check the viariable types:
rw.info()
```

```python
rw.describe().T
```

```python
#We see the two columns A_R2R2W_RollSpeed are equal to 0 all the values so we can delete this column:

rw.drop(['A_R2R2W_RollSpeed_Front','A_R2R2W_RollSpeed_Rear'], axis=1, inplace=True)
```

#### We want to analyse only the tests results:

```python
#Columns we want to analyse:
test_col = ['A_Indep_Front', 'A_RollSpeed_Front',
       'A_R2R2W_Front', 'KM_Front', 'KM_DEV_Front', 'A_Indep_Rear',
       'A_RollSpeed_Rear', 'A_R2R2W_Rear', 'KM_Rear', 'KM_DEV_Rear']
```

```python
#We plot the correlation matrix:
correlation = rw[test_col].corr()

mask = np.zeros_like(correlation, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(11, 9))

cmap = sns.diverging_palette(255,0,sep=77, as_cmap=True)

sns.heatmap(correlation, 
            mask=mask,
            cmap=cmap, 
            vmin=-1,
            vmax=1,
            center=0,
            linewidths=.5,
            annot=True,
            annot_kws={"size": 7}
           )
plt.show()
```

```python
#We analyse the two columns with 0.98 correlation:
rw[['KM_Front','KM_Rear']].describe()
```

```python
#We define a function to plot the main characterists of a variable:

def RW_eda(var):
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.40, 1)})

    # Plot
    sns.boxplot(rw[var][rw[var].notnull()], ax=ax_box)
    sns.distplot(rw[var][rw[var].notnull()], ax=ax_hist)
    plt.show()
```

```python
RW_eda('KM_Front')
```

```python
RW_eda('KM_Rear')
```

```python
#Scatter plot:
sns.scatterplot(x="KM_Front",y="KM_Rear",data=rw)
plt.show()
```

We can see that there is a group of tests results with big differences respect the main group of results.
We are going to analyse if this has an specific reason or htey are mistakes and we can redo the analysis without them.

```python
#Delete de outliers of KM_Front & KM_Rear in a new dataset to cheack again correlations.
#We separate the data without outliers and the outliers itself:

#KM_Front
Q1 = rw['KM_Front'].quantile(0.25)
Q3 = rw['KM_Front'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5* IQR
upper = Q3 + 1.5* IQR

no_outliers = rw.loc[(rw['KM_Front'] > lower) & (rw['KM_Front'] < upper)]
outliers1 = rw.loc[(rw['KM_Front'] < lower)]
outliers2 = rw.loc[(rw['KM_Front'] > upper)]
outliers = pd.concat([outliers1,outliers2])

#KM_Rear
Q1 = rw['KM_Rear'].quantile(0.25)
Q3 = rw['KM_Rear'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5* IQR
upper = Q3 + 1.5* IQR

no_outliers = no_outliers.loc[(no_outliers['KM_Rear'] > lower) & (no_outliers['KM_Rear'] < upper)]
outliers1 = rw.loc[(rw['KM_Rear'] < lower)]
outliers2 = rw.loc[(rw['KM_Rear'] > upper)]
outliers = pd.concat([outliers,outliers1,outliers2])
#Deleting the duplicate that can be included in the data:
outliers.drop_duplicates(keep=False,inplace=True) 
```

```python
#New correlation between the two columns without outliers:
no_outliers['KM_Front'].corr(no_outliers['KM_Rear'])
```

```python
#We start to analyse the outliers to check if has sense to delete it or not:

outliers.head()
```

```python
#Redefine the function:
def RW_eda(var):
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.40, 1)})

    # Plot
    sns.boxplot(no_outliers[var][no_outliers[var].notnull()], ax=ax_box)
    sns.distplot(no_outliers[var][no_outliers[var].notnull()], ax=ax_hist)
    plt.show()
```
